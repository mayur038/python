```python
import pandas as pd

# Sample DataFrame
data = {'x': [1, 2, 3, 4, 5], 'y': [2, 3, 4, 5, 6]}
df = pd.DataFrame(data)

# Compute covariance
covariance_df = df.cov().iloc[0, 1]

# Compute correlation
correlation_df = df.corr().iloc[0, 1]

print(f'Covariance (DataFrame): {covariance_df}')
print(f'Correlation (DataFrame): {correlation_df}')
```

```python
import numpy as np
from scipy.stats import skew,kurtosis
aa = np.random.randint(1,100,50)
skewness = skew(aa)
kurtosis = kurtosis(aa)

print(skewness)
print(kurtosis)
```
